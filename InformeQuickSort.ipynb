## 1. Problema de Ordenamiento

Un algoritmo de ordenamiento es un algoritmo que recibe un arreglo sin orden especifico y da como salida la secuencia dada anterior, pero con una relación de orden, es decir, el resultado de salida ha de ser un reordenamiento de la entrada que satisfaga una relación de orden (mayor a menor o viceversa).

El problema del ordenamiento ha atraído gran cantidad de atención, tal vez debido a que tener datos ordenados ayuda a una mejor comprensión o análisis de estos datos que si estuvieran sin ningún orden aparente.

**Entrada** arreglo de tamaño n: $a_{n-1}, a_1, a_5, . . . , a_2$

## 2. QuickSort

El QuickSort es un algoritmo creador por el científico británico en computación Charles Anthony Richard Hoare basándose en el paradigma divide y vencerás que permite en promedio ordenar n elementos en un tiempo n log n.

En el anterior informe vimos que había 3 formas de encontrar el máximo sub-arreglo, las 3 daban exactamente el mismo resultado, pero la forma de resolver el problema cambiaba, el QuickSort también tiene variaciones, pero estas variaciones radica en cuál pivote usar:

*   Ocupar el primer elemento del arreglo como pivote
*   Ocupar el último elemento del arreglo como pivote
*   Ocupar la posición que sea la mediana del arreglo como pivote
*   Ocupar un número random dentro del arreglo como pivote

### 2.1. Descripción del QuickSort

Como el MergeSort, el QuickSort es un algoritmo que ocupa el paradigma, divide y vencerás utilizando un pivote para comparar si es mayor o menor y a partir de ahí se crea cuál será la mitad izquierda y la mitad derecha para que después el problema se divida con recursividad.


Pseudocódigo del QuickSort:
```
//Funcion quicksort que recibe el arreglo[], posicion inicial, posicion final

quickSort(arreglo[], low, high) {

    if (low < high) {

        //Funcion partition que devuelve el pivote
        pi = partition(arreglo, low, high);
        
        //Recursividad
        quickSort(arreglo, low, pi – 1);  // izquierda

        quickSort(arreglo, pi + 1, high); // derecha
    }
}
```


Observando como se comporta este algoritmo se ve que hace lo siguiente:

*   Se retorna un pivote, y la función partition reordena un poco el arreglo
    en torno al pivote elegido.

*   Recursividad de la función QuickSort pero con la parte izquierda del arreglo

*   Recursividad de la función QuickSort pero con la parte derecha del arreglo

*   Esta recursividad termina cuando la posición de la derecha es más pequeño
    que el de la izquierda

Se puede observar que la función partition es la parte más importante del algoritmo debido a que se encarga de crear el pivote como a su vez reordenar el arreglo respecto a este pivote, como también se puede observar que el pivote no está incluido en ninguno de los 2 sub-arreglos.

### 2.2. Función Partition

Esta función se encarga de retornar un pivote y de paso reordenar el arreglo respecto a este pivote, en este caso explicaremos como funciona. 

**Mostrando 2 ejemplos** :


Pseudocódigo de la función Partition con pivote al último elemento del arreglo:

```
partition (arr[], low, high){

    // pivote (Elemento que esta mas a la derecha del arreglo)
    pivot = arr[high];  
    
    //Indice del elemento mas pequeño e indica la posicion
    //correcta del pivote hasta el momento
    i = (low – 1)

    for (j = low; j <= high- 1; j++){

        // si el elemento actual es mas pequeño que el pivote
        if (arr[j] < pivot){
            i++;  
            swap arr[i] and arr[j]
        }
    }

    swap arr[i + 1] and arr[high])
    return (i + 1)
}
```
Observando el pseudocódigo la función hace estos pasos:

1.  Elige un pivote (El último del arreglo)
2.  Índica el índice más pequeño que tiene el arreglo
3.  Un ciclo for j para recorrer el arreglo
     
     * En cada iteración se preguntará si el valor actual es menor que el pivote
          * Si es así se intercambia el valor arr[ j ] con la posición arr[ i ]
          * Se suma i++ 

4.  Se intercambia la posición arr[ i + 1 ] con la posición del pivote
5.  Retorna la posición del pivote


Pseudocódigo de la función Partition con pivote en la mitad del arreglo:

```
partition(A, start, end):
    
    //Indice menor 
    i = start - 1 
    
    //Pivote un numero random dentro del arreglo
    pivot = A[(start+end)//2]
    
    //Indice mayor
    j = end + 1

    while True:
        i+=1
        while (A[i] < pivot):
            i+=1 //se mueve el indice menor a la derecha

        j-=1
        while (A[j]> pivot):
            j-=1 //se mueve el indice mayor a la izquierda

        if i>=j:
            return j //retorna la posicion del pivote
        
        //swap
        A[i], A[j] = A[j], A[i] 
} 
```

Observando el pseudocódigo la función hace estos pasos:

1.  Índica el índice más pequeño que tiene el arreglo
2.  Elige un pivote (el número central)
3.  Índica el índice más grande que tiene el arreglo
4.  Un ciclo while (true)
     
     * En el primer while dura hasta que el valor de A[ i ] sea mayor   
       que el pivote

          * se le suma 1 a i

     * En el segundo while dura hasta que el valor de A[ j ] sea menor que el 
       pivote

          * se le resta 1 a j

     * Si i es mayor o igual que j retorna j que es la posición del pivote

     * Swap entre A[ i ] y A[ j ] 

La complejidad de la función partition es de $O(n)$

````python
import numpy as np

def medianOfThree(arr, left, right):
    mid = (left + right)//2
    promedio = (arr[left]+arr[mid]+arr[right])//3
    diferencia1 = abs(arr[left]-promedio)
    diferencia2 = abs(arr[mid]- promedio)
    diferencia3 = abs(arr[right]- promedio)

    if diferencia1<diferencia2:
        if diferencia1<diferencia3:
            return left
        else:
            return right
    else:
        if diferencia2<diferencia3:
            return mid
        else:
            return right                

def partitionMediana(array, low, high):
  n = 1
  temporal = medianOfThree(arr,low,high)
  (array[temporal], array[high]) = (array[high], array[temporal])

  pivot = array[high]

  print ("Elige un pivote :", colored(f"{pivot}","yellow"))
  
  i = low - 1
  print ("Indice mas pequeño :",i)
  
  for j in range(low, high):
    print(colored(f"\nIteracion :","green"), colored(f"{n}","green")) 
    n += 1
    print (array[j],"menor o igual",pivot,"?")  
    if array[j] <= pivot:
      print(str(array[:j])[1:-1], ",", colored(f"{str(array[j:j+1])[1:-1]}", "red"), ",", str(array[j+1:high])[1:-1], colored(f"{str(array[high:high+1])}", "yellow"))
      i = i + 1
      print ("\nSe le suma 1 a i y queda , i = ", i)
      (array[i], array[j]) = (array[j], array[i])
      print("Se hace un swap entre  arrray[i] =",array[i],"y array[j] =",array[j])
      print((colored(f"{str(array[:i+1])}","blue")), colored(f"{str(array[i+1:high])}", "red"), colored(f"{str(array[high:high+1])}", "yellow"))
    else:
         print(colored(f" Falso por lo que se sigue sin cambiar nada","cyan")) 
  
  print(colored(f"\nFinalmente :","green")) 
  print("Se intercambia el array[i + 1] =",array[i + 1],"con array[high] =",array[high])
  (array[i + 1], array[high]) = (array[high], array[i + 1])
  print((colored(f"{str(array[:i+1])}","blue")), colored(f"{str(array[i+1:i+2])}", "yellow"), colored(f"{str(array[i+2:])}", "red"))
  
  return i + 1                    


arr =  [ 20 , 70 , 90 , 10 , 40 , 30 , 50 , 60 ]

print ("Arreglo de entrada :", arr, "\n")
posicion = partitionMediana (arr, 0, len(arr) - 1)

print ("\nArreglo pasado por la funcion partition :", str(arr))
print("Sub-arreglo izquierda :", (colored(f"{str(arr[:posicion])}","blue")))
print("Sub-arreglo derecho :",(colored(f"{str(arr[posicion + 1:])}","red")))
print ("Posicion del pivote : ", posicion)
print ("pivote = ", (colored(f"{str(arr[posicion:posicion+1])}","yellow")))
````

## 3. Implementacion en python

Implementaremos QuickSort en python, utilizare de pivote el ultimo elemento del arreglo.

````python
comparaciones = 0

def partition3(array, low, high, verbose = False):
  global comparaciones
  n = 1
  pivot = array[high]
  
  if verbose == True:
      print ("\n-----------------------------Funcion Partition-------------------------------")
      print ("Elige un pivote :", colored(f"{pivot}","yellow"))
  
  i = low - 1
  if verbose == True:
      print ("Indice mas pequeño :",i)
  
  for j in range(low, high):
    comparaciones += 1
    if verbose == True:
          print(colored(f"\nIteracion :","green"), colored(f"{n}","green")) 
          n += 1
          print (array[j],"menor o igual",pivot,"?")  

    if array[j] <= pivot:
      if  verbose == True:
          print(str(array[:j])[1:-1], ",", colored(f"{str(array[j:j+1])[1:-1]}", "red"), ",", str(array[j+1:high])[1:-1], colored(f"{str(array[high:high+1])}", "yellow"))

      i = i + 1

      if verbose == True:
          print ("\nSe le suma 1 a i y queda , i = ", i)

      (array[i], array[j]) = (array[j], array[i])

      if verbose == True:
          print("Se hace un swap entre  arrray[i] =",array[i],"y array[j] =",array[j])
          print((colored(f"{str(array[:i+1])}","blue")), colored(f"{str(array[i+1:high])}", "red"), colored(f"{str(array[high:high+1])}", "yellow"))

    else:
        if verbose == True:
             print(colored(f" Falso por lo que se sigue sin cambiar nada","cyan")) 
  
  if verbose == True:
     print(colored(f"\nFinalmente :","green")) 
     print("Se intercambia el array[i + 1] =",array[i + 1],"con array[high] =",array[high])

  (array[i + 1], array[high]) = (array[high], array[i + 1])

  if verbose == True:
     print((colored(f"{str(array[:i+1])}","blue")), colored(f"{str(array[i+1:i+2])}", "yellow"), colored(f"{str(array[i+2:])}", "red"))
     print ("<\n----------------------------Fin Funcion Partition----------------------------------")
  
  return i + 1


def quick_sort(array, low, high, verbose = False):
    global comparaciones
    comparaciones += 1
    if verbose == True:
           print("\n-------------Funcion QuickSort------------ ")
           print("Arreglo de entrada :",str(array))

    if low < high:
       pi = partition3(array, low, high, verbose)

       if verbose == True:
           print ("\nArreglo entregado por la funcion partition :", str(array))
           print("Sub-arreglo izquierda :", (colored(f"{str(array[:pi-1])}","blue")))
           print("Sub-arreglo derecho :",(colored(f"{str(array[pi + 1:])}","red")))

       quick_sort(array, low, pi - 1, verbose)

       quick_sort(array, pi + 1, high, verbose)

	
		
#Ejemplo
array = [ 10, 7, 8, 9, 1, 5, 2, 4]

print(f'Arreglo de entrada : {array}')
print("\n------------------------------")
quick_sort(array, 0, len(array) - 1)
print(f'\nArreglo arreglado por el QuickSort : {array}')
````

A través de esto se puede observar que el pivote queda al final o al principio del arreglo, provocando que el caso no se divida en parte como el paradigma lo indica, creando el siguiente árbol recursivo:

perocaso.png

Esto significa que de $n$ veces que se divide el arreglo en cada nodo se hará 1 vez la función partition que tiene de complejidad $O(n)$ obteniendo la siguiente sumatoria.


$$\sum_{n=1}^{n}n= \frac{n}{2} \cdot(n + 1)$$

Si calculamos la complejidad del resultado nos quedará que:


$$\frac{n}{2} \cdot(n + 1) = O( n^2 )$$ 

Por lo que el **tiempo de ejecución en el peor caso es de $O(n^2)$**

**Caso promedio:**
Es el caso más común y la razón por la que se prefiere este algoritmo a pesar de tener ese problema con los arreglos ordenados.

Para un caso promedio tomaremos, por ejemplo un caso en particular, un arreglo tamaño n donde cada vez que se usa la función partition esta lo parte exactamente por la mitad.

mejorcaso.png



En cada fase del árbol se hará una comparación n incluso a medida que iremos bajando del árbol, esta n comparaciones son por parte del tamaño del arreglo, así que para determinar el total de comparaciones en todo el árbol hay que multiplicar $n*altura$ donde $altura = log(n)$.

Por lo que el **tiempo de ejecución promedio del QuickSort es de $O( n log(n) )$.**

### 5.2. Correctitud del QuickSort


El algoritmo QuickSort genera un arreglo de salida: $[a_1, a_2, . . . , a_n]$, exactamente con los mismos valores del arreglo de entrada ordenados de menor a mayor, es decir, $a_1\leq a_2 \leq . . .  \leq a_n$.

Para probar la *correrctitud* del QuickSort usare Inducción.

Para probar que un $T(n)$, para todos lo números naturales n es correcto, debemos:


1.   Probar $T(n)$ para un caso base, por ejemplo $T(1)$
2.   Probar que si $T(m)$ es cierto para $m<n$, entonces $T(n)$ también lo es.

Nuestra hipótesis $T(n)$ es que QuickSort, ordena un **arreglo** con la secuencia de $n$ elementos de entrada ordenados de menor a mayor.

**Caso Base:**

Cuando $n = 1$, QuickSort no realiza ninguna operación simplemente no le hace nada al arreglo.

**Paso inductivo:**

Para $n$ elementos, partition utiliza un pivote para determinar los 2 sub-arreglos de tamaño $m_1 < n$ y $m_2 < n$. Como a su vez reorganiza el arreglo en torno al mismo pivote, después QuickSort emplea lo 2 sub-arreglos para volver a ejecutar el mismo proceso. Como el tamaño de los sub-arreglos es menor a $n$, con el paso inductivo asumimos que son ordenadas correctamente.

La reorganización de los arreglos las efectúa la función partition que como vimos en el punto 5.1. Es verdadero por lo que QuickSort también es verdadero.

## 6. Experimentos

### 6.1. Tiempo de ejecución

Para realizar la comparación de tiempo de ejecución, se comparará con 2 algoritmos de ordenamiento que realizan el mismo trabajo que QuickSort, pero de manera distinta, estos algoritmos son MergeSort y HeapSort.

MergeSort:

````python
def mergeSort(arr):
	if len(arr) > 1:

		mid = len(arr)//2

		L = arr[:mid]
		R = arr[mid:]

		mergeSort(L)
		mergeSort(R)

		i = j = k = 0
		while i < len(L) and j < len(R):
			if L[i] < R[j]:
				arr[k] = L[i]
				i += 1
			else:
				arr[k] = R[j]
				j += 1
			k += 1

		while i < len(L):
			arr[k] = L[i]
			i += 1
			k += 1

		while j < len(R):
			arr[k] = R[j]
			j += 1
			k += 1
````

HeapSort:

````python
def heapify(arr, N, i):
	largest = i
	l = 2 * i + 1	
	r = 2 * i + 2	

	if l < N and arr[largest] < arr[l]:
		largest = l


	if r < N and arr[largest] < arr[r]:
		largest = r


	if largest != i:
		arr[i], arr[largest] = arr[largest], arr[i] 

		heapify(arr, N, largest)


def heapSort(arr):
	N = len(arr)

	for i in range(N//2 - 1, -1, -1):
		heapify(arr, N, i)

	for i in range(N-1, 0, -1):
		arr[i], arr[0] = arr[0], arr[i] 
		heapify(arr, i, 0)
````

Estos algoritmos tienen en común con QuickSort que su complejidad promedio es $O(nlog(n))$, por lo que compararemos cuál de los 3 es más rápido incluso si tiene la misma complejidad.

Para este experimento se usará tamaño de arreglo desde 5 hasta 500 con números al azar desde el 1 hasta 999.

````python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat
from copy import deepcopy
import random

x=[]
y_quick = []
y_merge = []
y_heap = []

for n in range(5,501):

  a = random.sample(range(1, 1000), n)
  b = deepcopy(a)
  c = deepcopy(a)

  t_quick = repeat(setup="from __main__ import quick_sort", stmt=f"quick_sort({a},{0},{len(a)-1})", repeat=1, number=10)
  t_merge = repeat(setup="from __main__ import mergeSort", stmt=f"mergeSort({b})", repeat=1, number=10)
  t_heap = repeat(setup="from __main__ import heapSort", stmt=f"heapSort({c})", repeat=1, number=10)

  x.append(n)
  y_quick.append(t_quick)
  y_merge.append(t_merge)
  y_heap.append(t_heap)


plt.plot(x,y_quick)
plt.plot(x,y_merge)
plt.plot(x,y_heap)

plt.legend(["QuickSort", "MergeSort","HeapSort"])

plt.xlabel('n')
plt.ylabel('Tiempo en ms')
plt.show()
````

### Analisis del grafico

Se puede ver como el QuickSort es más rápido aunque por milésimas de segundo respecto a los otros 2 algoritmos, esto puede ser debido a que el heapSort por su parte realiza una buena cantidad de acciones para ordenar, pasos extra que en comparación con el QuickSort como el reordenamiento del árbol binario, mientras que en el caso del MergeSort puede ser debido a que su función Merge Solo junta sub-arreglos mientras que partition reorganiza los sub-arreglos respecto a un pivote ocasionando que una vez que se llegue al caso donde no se llama a sí misma la función QuickSort esta ya termino su trabajo mientras que MergeSort requiere llegar al caso base y a partir de ahí se empieza a devolver en las recursiones para terminar de ordenar el arreglo.
