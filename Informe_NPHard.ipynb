# Informe sobre el conjunto NP-Hard

## 1. ¿ Qué quiere decir que un problema NP-Hard? 
**NP-Hard es un conjunto de problemas tan difíciles como los del conjunto NP-Completo, pero no todos los problemas NP-Hard están en NP**, estos problemas se caracterizan por ser intratables, o sea que su tiempo de ejecución excede en complejidad a la polinómica.

Sí definimos el tiempo de ejecución de un problema NP-Hard como $O(F(n))$ entonces:

$$F(n) > n^k, k ∈ \mathbb N$$

**$F(n)$ crece más rápido que cualquier función polinómica**, funciones como la factorial o las exponenciales.

Su definición formal es: Un problema $A$ es NP-Hard, si existe un problema NP-Completo $B$, tal que $B$ es reducible al problema $A$ en tiempo polinomial. 

## 2. ¿ Qué quiere decir que $P\neq NP$ ?
Los problemas se clasifican según su complejidad: 

Suponiendo lo siguiente : $k ∈ \mathbb N$ y $n$ es el tamaño del problema: 

+ Los problemas que se pueden resolver en un tiempo polinomial $O(n^k)$ o menor, estos problemas pertenecen al conjunto $P$

+ Los problemas que se les puede verificar si una  solución es correcta en tiempo polinomial $O(n^k)$ o menor, el conjunto descrito se le llama $NP$.

Podemos decir que todos los problemas en $P$ están en $NP$ porque si tenemos una solución en tiempo polinómica podremos saber también en tiempo polinómico si nuestra solución es correcta. ¿Pero al revés se cumple?, Si podemos verificar que una solución es correcta en tiempo polinómico, ¿implicaría que podemos encontrar una solución en tiempo polinómica?, la respuesta hasta el día de hoy no tiene una solución clara.

¿Pero como puede ser eso?, que $P \subseteq NP$ pero no al revés, para explicar esto daremos un ejemplo sencillo de conjuntos:

Todos conocemos el conjunto Natural $\mathbb N$, este conjunto pertenece a otro Conjunto más grande, el de los reales $\mathbb R$. Si tomamos un elemento cualquiera del conjunto Natural, este elemento siempre pertenecerá a los reales, pero si tomamos un elemento en los reales no siempre pertenecerá a los naturales, como por ejemplo una fracción.

Esto implica que para el conjunto $NP$ sea igual al conjunto $P$ todos los problemas en $NP$ tienen que estar en $P$ o sea que puedan resolverse en tiempo polinómico. Los problemas pertenecientes a los conjuntos NP-Hard y NP-Completo son los que niegan esta implicación $P = NP$, así que hasta que no se muestre lo contrario o se resuelve algún problema NP-Hard en tiempo polinómico, todos dan por verdadero la conjetura de que $P\neq NP$.

## 3. ¿ Cómo se prueba que un problema es NP-Hard usando reducciones?

### Reducción
Un problema $A$ se puede reducir a un problema $B$, si los resultados de $B$ se pueden traducir en una que resuelva $A$.

ejemplo2.png

$A$ puede ser resuelta usando una cantidad soluciones de $B$, como las vistas en programación dinámica, donde para resolver un problema se resolvían una serie de subproblemas y también puede realizarse un trabajo adicional que implica trabajar con todas las soluciones de $B$ para traducirlas en una solución de $A$.

### Como usar reducción para probar que un problema es NP-Hard
Suponga que el problema $Α$ se reduce a otro problema $ß$.

Suponga que $Α$ es NP-hard, lo que significa que encontrar un algoritmo polinomial para $A$ refutaría la conjetura P ≠ NP. Pero si también existiera un algoritmo polinomial para resolver $ß$ también implicaría en $A$ (porque $A$ se reduce en $ß$) y eso también refuta la conjetura P ≠ NP, en otras palabras $ß$ también es NP-hard.

Por lo que si un problema $A$ se reduce a un problema $ß$ y $A$ es NP-hard entonces $ß$ también lo es.

#### **Como probar que un problema es NP-Hard**:

Si queremos probar que $ß$ es NP-Hard:
1. Elegir un problema $A$ que sea NP-Hard.
2. Probar que el problema $A$ se reduce a $ß$.

**Pre-proceso**: Teniendo un grafo no dirigido $G = (V,E)$, un nodo inicial $s$ y un nodo final $t$, en el pre-proceso tenemos que traducir la entrada del problema UHP para que pueda funcionar dentro del problema TSP, para eso, primero se cierra la brecha entre las rutas que visitan todos los nodos y los ciclos que visitan todos los nodos al agrandar $G$ con un nodo adicional $v_0$ y arcos que conectan $v_0$ con $s$ y $t$, luego se asigna a todos los arcos del grafo agrandado un coste igual a 0. Para completar la instancia para TSP tenemos que agregar los arcos faltantes para completar el grafo (para formar un grafo $G'$ con un conjunto de arcos $V ∪ \{{v_0}\}$) y a estos arcos asignarle un coste igual a 1.

## 5. Una solución posible para resolver el problema

### Cuantas rutas posibles hay en el TSP
Esto es una permutación sin repetición, las posibles ordenaciones de un conjunto de $n$ nodos sin que se repita ningún nodos en una combinación, el valor de una permutación sin repetición se representa por un factorial.

En el problema hay $n!$ rutas posibles, aunque esto se puede simplificar un poco, debido a que como empezamos desde un nodo en específico reduce el número de rutas, siendo esto $(n-1)!$, como también no importa la dirección en la que el vendedor ambulante se mueve lo que también reduce las rutas en la mitad por lo que finalmente queda $(n-1)!/2$ rutas posibles.

### Posible solución
Una forma de resolverlo, aunque posiblemente sea la peor, es el de calcular cada ruta posible, eso significa que tendríamos que calcular $n!$ rutas para $n$ nodos, así que podríamos por lo menos reducir un poco ese $n!$ y la única manera de hacerlo buscar una forma efectiva de calcular combinaciones que sean candidatas a solución.

#### Como encontraríamos una ruta válida sin tener que verlas todas
Intentaremos hacer una elección greedy, tomando un nodo cualquiera como el primer nodo y a partir de ahí elegiremos los siguientes nodos basándonos en los nodos vecinos y tomaremos el que contenga el arco más pequeño (sin contar el nodo inicial y los arcos que unen el nodo con el nodo inicial) una vez elegido el nodo vecino se repite el proceso hasta que no haya ningún nodo vecino, después a esta ruta uniremos el último nodo de la ruta y lo conectaremos con el nodo inicial usaremos el mismo proceso pero con el primer nodo también uniéndolo con el nodo inicial.

A medida que vayamos creando la ruta iremos sumando los arcos y también cuando unamos la ruta con el nodo inicial, este valor se comparara con los valores de rutas anteriores, siempre guardando la ruta con el menor valor.

En cuanto al tiempo de ejecución vemos que dentro del paso a paso primero tenemos un while, este while recorrerá los elementos de la cola de prioridad que es la cantidad de nodos - 1 (excepto el nodo inicial), después tenemos un for que recorre la cantidad de nodos - 2(excepto el nodo inicial y el primer nodo elegido) para finalizar con otro nodo que recorrerá cada arco del grafo completo por lo que podríamos obtener lo siguiente:
